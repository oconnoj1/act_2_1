## Background on KDE

KDE stands for Kernal Density Estimation. This is a plot that is used to estimate the probability density function (pdf) for a given data set. This plot is very similar to a histogram, but instead of grouping observations in bins, this plot is drawing a line over the points. A KDE plot would fit over a histogram very well. This plot is not making any assumptions about how the data is distributed, but rather it is just drawing an estimated distribution based on how frequently an observation appears. 



KDE works by creating one kernel for each point in the data set. You can use whatever you would like as the kernel, but the most common and default in seaborn is to use the normal kernel. Then it adds each of these kernels together, and creates an esimtated pdf for the observed data. See this image below for an example. This example is using the normal kernel.

Seaborn allows us to create both a univariate and bivariate KDE plot. We will cover both of these, starting with the univariate KDE plot. We will be using the cars dataset, as it comes loaded with seaborn.

### Univariate

Univariate KDE plots are really straightforward. The syntax is as follows.

sns.**kdeplot**(df.column_name);

This code will produce a KDE plot for the variable in the data frame.

Additionally, there are several options to keep in mind if you would like to customize your plot.

* fill = True - Shade in the plot
* Color = (color) - Colors the line of the KDE plot
* bw_adjust = (number) -  Bandwidth. Bandwidth allows to change how smooth vs jagged your KDE plot is. Higher badnwidth leads to smoother plots.
* cumulative = True - creates a cummulative distribution function (cdf)
* More customization options can be found in seaborn.kdeplot documentation.

```python
sns.kdeplot(cars.horsepower, bw_adjust = 1);
```

```python
sns.kdeplot(cars.acceleration, fill = True, color = 'red');
```

### Bivariate

A bivariate KDE plot shows the estimated joint pdf between two variables. This plot creates a contour map, where each contour represents higher value of the joint pdf. This iamge below shows what a joint KDE plots is trying to estimate. Think about each contour as a different color in this iamge.



The code to create a joint KDE is really similar to the univariate KDE plot. Insteaf of only passing 1 variable, you should pass multiple.

sns.kdeplot(x = (x_var), y = (y_var));

Here are some additional arguments you can give to the kdeplot function:
* fill = True - Shades in the contours. This is useful, since you can see which values have the same value.
* Many of the same arguments as the univaraite since they are the same function!

```python
sns.kdeplot(x = cars.acceleration, y = cars.horsepower, fill = True);
```
